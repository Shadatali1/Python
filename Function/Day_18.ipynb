# **Higher Order Functions**

### Concept Explanation

👉 A Higher Order Function is a function that:

Takes another function as input\
OR\
Returns another function

💡 Simple:\
👉 “Function working with another function”

### Story Time

Think of a boss 👨‍💼

Boss gives task to worker\
Worker does job

👉 Boss = Higher Order Function\
👉 Worker = Normal Function

### PART 1 – Function as Argument

```python
def greet(name):
    return "Hello " + name
```

```python
def process(func):
    print(func("Ali"))
```

#### **Run**

```python
process(greet)
```

💡 We passed a function as input 🔥

### PART 2 – Function Returning Function

```python
def outer():

    def inner():
        return "I am inner function"

    return inner
```

#### **Run**

```python
result = outer()
print(result())
```

💡 Function returned another function 😎

### PART 3 – Real Example

```python
def calculator(operation):

    def add(a, b):
        return a + b

    def multiply(a, b):
        return a * b

    if operation == "add":
        return add
    else:
        return multiply
```

#### **Run**

```python
func = calculator("add")
print(func(2, 3))
```

### Lambda + Higher Order

```python
def apply(func, value):
    return func(value)
```

```python
print(apply(lambda x: x * 2, 5))
```

### Practice Tasks

```python
# Task 1:
def square(x):
    return x * x

def apply_func(func, value):
    return func(value)

# Call square using function
result = apply_func(square, 5)

print(result)
```

```python
# Task 2:
def message():
    return "Hello Student"

def print_message(func):
    print(func())

# Pass message function
print_message(message)
```

```python
# Task 3:
def choose(op):
    def add(a, b):
        return a + b

    def subtract(a, b):
        return a - b

    if op == "add":
        return add
    elif op == "subtract":
        return subtract
```

```python
operation = choose("add")
print(operation(10, 5))   # 15

operation = choose("subtract")
print(operation(10, 5))   # 5
```
