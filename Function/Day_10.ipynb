### Nested Functions (Function inside Function)

**Simple Meaning**

👉 A function inside another function

💡 One function can use another function inside it

### Story Time

Think of a company 🏢

- Manager 👨‍💼 → main function
- Worker 👷 → inner function

Manager tells worker what to do

💡 Worker (inner function) works inside manager only

### Python Example

```python
def outer():
    print("I am outer function")

    def inner():
        print("I am inner function")

    inner()
```

```python
outer()
```

### Important Rule

❌ You CANNOT call inner function outside

inner()   # ERROR ❌

💡 Because inner function is hidden inside

### Real Example

```python
def calculator(a, b):

    def add():
        return a + b

    def multiply():
        return a * b

    print("Sum:", add())
    print("Multiply:", multiply())
```

```python
calculator(5, 3)
```
