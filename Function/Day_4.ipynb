#  **Return Values (Function gives something back)**

**Simple Meaning**\
Parameter/input → you give to function\
Return → function gives something back

Think: function is a machine 🏭

Input goes in → machine processes → output comes out

### Story Time

Tea machine example ☕:

You press button → machine gives tea cup\
You can drink it, give to friend, or sell it

💡 Output = return

### Python Example

```python
def add(a, b):
    return a + b
```

- return a + b → function gives the result back
- Not just printing

**Using the Return Value**

```python
result = add(5, 3)
print("Result is:", result)
```

### Difference: print vs return

```python
def add_print(a, b):
    print(a + b)   # only shows result

def add_return(a, b):
    return a + b   # gives result for later use
```

**Try**

```python
x = add_print(2, 3)   # prints 5, but x = None
y = add_return(2, 3)  # y = 5
print(x, y)
```
