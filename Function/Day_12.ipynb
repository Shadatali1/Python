# **Lambda Functions (Short Functions)**

**Simple Meaning**

👉 Lambda = small one-line function

💡 No need to write full def

### Story Time

Normal function = cooking full meal 🍛\
Lambda = instant noodles 🍜

👉 Fast, simple, quick

**Normal Function**

```python
def add(a , b):
    return a + b
```

**Lambda Function**

```python
add = lambda a ,b: a + b

# Use Lambda
print(add(3 , 4))
```

### Structure

lambda parameters : expression

```python
lambda x: x * 2
```

### More Examples
**1️⃣ Square**

```python
square = lambda x: x * x
square(6)
```

**2️⃣ Multiply**

```python
multiply = lambda a , b: a * b
multiply(4 , 5)
```

### When to Use Lambda?

✔ Small work\
✔ One-line logic\
✔ Quick calculations\
✔ Used in AI / Data Science (map, filter)

### Limitations

❌ Only one expression\
❌ Not for big logic\
❌ Less readable sometimes
