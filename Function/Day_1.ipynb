# **What is a Function?**

A function is like a machine 🤖\
👉 You give it a task → it does the job

### Story (Very Important)

Imagine you have a tea shop

Every time a customer comes, you:

- boil water
- add tea
- add sugar
- serve

This is repeating work again and again\
💡 So you make a tea machine

**Now:**\
👉 Press button → tea ready

That machine = function

### Python Example

```python
def make_tea():
    print("Tea is ready ☕")
```

**Understand line by line:**
- def → means "define function"
- make_tea → function name
- () → inputs (empty for now)
- : → start function
- print() → what it does

### Calling the Function

Function will NOT run automatically ❌\
You must call it

```python
make_tea()
```

### Real Life Examples

| Real Life       | Function      |
| --------------- | ------------- |
| TV remote       | call function |
| Washing machine | function      |
| Calculator      | function      |


### Common Mistakes

❌ Forget to call function\
❌ Wrong indentation\
❌ Missing ()
