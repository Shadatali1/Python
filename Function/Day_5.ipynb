# **print() vs return (Deep Understanding)**

### Simple Difference

| print() 🖨️  | return 📦           |
| ------------ | ------------------- |
| Shows output | Gives output back   |
| Cannot reuse | Can reuse           |
| Ends there   | Sends value outside |


### Story Time

Imagine:

👨‍🍳 Chef in kitchen

**Case 1: print()**

Chef just shows food 👀
👉 “Look, food is ready!”

**Case 2: return**

Chef packs food and gives you 🍱\

- eat

💡 That’s return power

### Example 1 (print)

```python
def add(a, b):
    print(a + b)
    
result = add(2, 3)
print(result)
```

**Why None?**\
Because function didn’t return anything

### Example 2 (return)

```python
def add(a, b):
    return a + b
result = add(2, 3)
print(result)
```

### Golden Rule

👉 If you want to use result later → use return\
👉 If you just want to show → use print

### Real Data Science Thinking

In AI/Data Science:

You must use `return`\
Because data moves from one function to another
