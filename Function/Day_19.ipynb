# **map(), filter(), reduce()**

👉 These are Higher Order Functions

They:

- take a function
- apply it to data

💡 Very useful for working with lists / datasets

### PART 1 – map() (Apply function to all items)

**Story**

Teacher gives same homework to all students 📚\
👉 Same task → all students

💡 That is map()

```python
numbers = [1, 2, 3, 4]
```

```python
square = lambda x: x * x
```

```python
result = map(square, numbers)
print(list(result))
```

🔅 Apply function to each item

### filter() (Select some items)

```python
numbers = [1, 2, 3, 4, 5, 6]
even = lambda x: x % 2 == 0
result = filter(even, numbers)
print(list(result))
```

### reduce() (Combine values)

```python
from functools import reduce
numbers = [1, 2, 3, 4]
result = reduce(lambda a, b: a + b, numbers)
print(result)
```

```python
# combine Example
numbers = [1, 2, 3, 4, 5, 6]

even = filter(lambda x: x % 2 == 0, numbers)
square = map(lambda x: x * x, even)

print(list(square))
```

### Practice Tasks 🎯
✅ **Task 1 – Map**\
nums = [2, 4, 6]

👉 Double all numbers using map

✅ **Task 2 – Filter**\
nums = [10, 15, 20, 25]

👉 Get numbers > 15

✅ **Task 3 – Reduce**\
nums = [1, 2, 3, 4]

👉 Find product (multiply all)

```python
nums = [2,4,6]
# Double all numbers using map

result = map(lambda x: x * 2, nums)
print(list(result))
```

```python
# Task 2 – Filter
nums = [10, 15, 20, 25]

# Get numbers > 15
result = filter(lambda x: x > 15 , nums)
print(list(result))
```

```python
# Task 3 – Reduce
nums = [1, 2, 3, 4]

# Find product (multiply all)
from functools import reduce

result = reduce(lambda a , b: a*b , nums )
print(result)
```

### Common Mistakes

❌ Forget list() with map/filter\
❌ Forget from functools import reduce\
❌ Confusing map vs filter

### Mini Challenge 🤖
nums = [5, 10, 15, 20]

👉 Do:

Filter numbers > 10\
Double them\
Sum all

```python
nums = [5, 10, 15, 20]
greator_then_10 = filter(lambda x:x > 10, nums)
doubles = map(lambda x:x * 2 , greator_then_10)
from functools import reduce
all_sum = reduce(lambda a,b:a+b, doubles)
print(all_sum)
```
