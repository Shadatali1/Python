# ***args (Multiple Inputs)**

**Simple Meaning**

👉 *args allows a function to take unlimited number of inputs

💡 You don’t need to fix number of parameters

### Story Time

You open a restaurant 🍽️

Customers order:

1 item\
3 items\
10 items

👉 You don’t know how many items will come

💡 So you say:\
👉 “Give me all items” → *args

### Basic Example

```python
def total(*numbers):
    print(numbers)
```

**Try it**

```python
total(1, 2, 3)
total(5, 10, 15, 20)
```

`*args stores values in a tuple`

### Real Use – Sum All Numbers

```python
def total(*numbers):
    total = 0
    for i in numbers:
        total += i
    return total
```

```python
total(2,3,4,5,6,7)
```

### Mix with Normal Parameters

```python
def student(name, *marks):
    print("Name:", name)
    print("Marks:", marks)
```

```python
student("Ali", 80, 70, 90)
```

**Important Rule**

👉 *args must come after normal parameters

❌ Wrong:

def test(*args, name):

✅ Correct:

def test(name, *args):

### Practice Task
