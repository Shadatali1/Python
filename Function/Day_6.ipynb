# **Default Parameters (Easy but Powerful)**

**Simple Meaning**

👉 Default parameter = value already set

If user does NOT give input → function uses default value

### Story Time

You go to tea shop ☕

Shopkeeper says:\
👉 “By default, I make tea with 2 spoons sugar”

If you say nothing → you get 2 spoons 🍬🍬\
If you say “no sugar” → he changes it

💡 That default setting = default parameter

### Python Example

```python
def make_tea(sugar=2):
    print("Tea with", sugar, "spoons sugar ☕")
```

**Calling Function**
- Case 1: No input

```python
make_tea()
```

- Case 2: With input

```python
make_tea(1)
```

### Real Life Examples

| Situation           | Default |
| ------------------- | ------- |
| Phone brightness 📱 | medium  |
| AC temperature ❄️   | 24°C    |
| App settings        | pre-set |

### More Example

```python
def greet(name="Guest"):
    print("Hello", name)
greet()
greet("Ali")
```

### Important Rule

👉 Default parameter ALWAYS at the end

❌ Wrong:

```python
def test(a=5, b):
```

✅ Correct:

```python
def test(a, b=5):
```
