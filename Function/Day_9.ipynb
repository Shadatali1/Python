# **Multiple Return Values**

**Simple Meaning**

👉 A function can return more than one value

💡 Not only one result — many results at same time

### Story Time

You go to shop 🛒

You buy:

- Bread 🍞
- Milk 🥛

Shopkeeper gives both together

👉 Not one by one → multiple return

### Python Example

```python
def calc(a, b):
    return a + b, a - b
```

**Calling Function**

```python
result = calc(10, 5)
print(result)
```

**Python returns a tuple**

### Better Way (Unpacking)

```python
sum_result, sub_result = calc(10, 5)

print("Sum:", sum_result)
print("Sub:", sub_result)
```

🔅 This is very important

### Real Life Example

| Situation   | Multiple Output   |
| ----------- | ----------------- |
| ATM         | balance + receipt |
| Exam result | marks + grade     |
| App login   | status + message  |

### More Example

```python
def student(name, marks):
    if marks >= 50:
        return name, "Pass"
    else:
        return name, "Fail"
```

```python
n, result = student("Ali", 80)
print(n, result)
```
