# ****kwargs (Named Unlimited Inputs)**

### Concept Explanation (Teacher Talking 👨‍🏫)

👉 Yesterday we learned *args\
It takes many values (no names)

👉 Today:\
**kwargs takes many values WITH names

### Story Time

You run an online store 🛒

Customer sends order like:

name = Ali\
item = Shoes\
price = 2000

👉 This is named data

💡 You don’t know how many details customer sends\
So you use:

👉 **kwargs = collect all named data

### Basic Python Example

```python
def show_info(**data):
    print(data)
```

```python
show_info(name="Ali", age=20, city="Lahore")
```

💡 It stores data in dictionary

### Loop Through Data

```python
def show_info(**data):
    for key, value in data.items():
        print(key, ":", value)
```

```python
show_info(name="Ali", age=20, city="Lahore")
```

### Real Example (Student Data)

```python
def student(**info):
    print("Student Details:")
    for key, value in info.items():
        print(key, ":", value)
```

```python
student(name="Ali", marks=85, grade="A")
```

| Feature | *args   | **kwargs    |
| ------- | ------- | ----------- |
| Type    | Tuple   | Dictionary  |
| Input   | Values  | Key + Value |
| Example | (1,2,3) | name="Ali"  |


### Using Both

```python
def demo(*args, **kwargs):
    print("Args:", args)
    print("Kwargs:", kwargs)
```

```python
demo(1, 2, 3, name="Ali", age=20)
```

### Important Rules

✔ *args comes before **kwargs

✔ Names must be written:

```python
name="Ali"
```

❌Wrong

```python
"Ali"
```
