# **Parameters (Input to Function)**

**Simple Meaning**

👉 Parameter = Input you give to function

### Story Time (Very Important)

You go to tea shop:

Shopkeeper asks:

Sugar kitni? 🍬\
Milk kitna? 🥛

Every customer wants different tea

💡 So machine must accept input

### Python Example

```python
def make_tea(sugar):
    print("Tea with", sugar, "spoons sugar ☕")
```

**👉 Here:**

- sugar = parameter (input)

### Calling Function with Input

```python
make_tea(2)
make_tea(5)
```

`Same function → different result 🔥`

### Real Life Example

| Function   | Input   |
| ---------- | ------- |
| Calculator | numbers |
| ATM        | amount  |
| Food order | items   |

### Practice Time
