### Step 1 – Basic Function

```python
def student_result(name , marks):
    if marks >= 50:
        return name + " Pass"
    else:
        return name + " Fail"
```

```python
print(student_result("Ali", 80))
print(student_result("Ahmed", 40))
```

### Step 2 – Add Grade System

```python
def student_result(name, marks):
    if marks >= 80:
        grade = "A"
    elif marks >= 60:
        grade = "B"
    elif marks >= 50:
        grade = "C"
    else:
        grade = "Fail"
    return name + " got grade " + grade
```

```python
print(student_result("Ali", 85))
print(student_result("Sara", 65))
print(student_result("Ahmed", 45))
```

### Step 3 – Use Lambda Inside

```python
def student_result(name, marks):

    grade = lambda m: "A" if m >= 80 else \
                      "B" if m >= 60 else \
                      "C" if m >= 50 else "Fail"

    return name + " got grade " + grade(marks)
```

### Step 4 – Add Default Value

```python
def student_result(name="Unknown", marks=0):

    grade = lambda m: "A" if m >= 80 else \
                      "B" if m >= 60 else \
                      "C" if m >= 50 else "Fail"

    return name + " got grade " + grade(marks)
```

```python
print(student_result())
print(student_result("Ali"))
print(student_result("Ali", 90))
```

### Step 5 – Advanced Version (Optional 🔥)

👉 Return multiple values:

```python
def student_result(name, marks):

    grade = "A" if marks >= 80 else \
            "B" if marks >= 60 else \
            "C" if marks >= 50 else "Fail"

    return name, marks, grade
```

```python
n, m, g = student_result("Ali", 85)
print(n, m, g)
```
