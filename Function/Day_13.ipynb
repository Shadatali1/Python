### Part 2 – Mini Project 1: Smart Calculator 🔢

👉 Mix normal + lambda

```python
add = lambda a, b: a + b
sub = lambda a, b: a - b
mul = lambda a, b: a * b
div = lambda a, b: a / b
```

```python
print(add(5, 3))
print(sub(5, 3))
print(mul(5, 3))
print(div(5, 3))
```

### Part 3 – Mini Project 2: Robot Messages

```python
robot = lambda name, task: name + " is doing " + task
```

```python
print(robot("Ali", "cleaning"))
print(robot("AI", "learning"))
```

### Part 4 – Mini Project 3: Student Result System

```python
result = lambda marks: "Pass" if marks >= 50 else "Fail"
```

```python
print(result(80))
print(result(40))
```

### Final Challenge (Important 🔥)

👉 Combine everything:

```python
def student(name, marks):
    result = lambda x: "Pass" if x >= 50 else "Fail"
    return f"{name}  {result(marks)}"
```

```python
print(student('Shadat Ali' , 40))
```
