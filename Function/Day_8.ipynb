# **Keyword Arguments (Smart Way to Pass Values)**

**Simple Meaning**\
👉 Normally we give values by position

```python
add(2, 3)
```

👉 But with keyword arguments, we give by name

```python
add(a=2, b=3)
```

💡 `This makes code clear and safe`

## Story Time

You order food 🍔

Without keyword:

👉 “Give me item 1, item 2”\
Confusing 😵

With keyword:

👉 “Burger = 1, Drink = 1”\
Clear ✅

💡 That’s keyword arguments

### Python Example

```python
def student(name, age):
    print(name, "is", age, "years old")
```

**Normal Call**

```python
student("Ali", 20)
```

**Keyword Call**

```python
student(name="Ali", age=20)
```

✅Same Output, But more readable

### Powerful Feature

👉 You can change order

```python
student(age=20, name="Ali")
```

💡 Still works

### Important Rule

❌ Wrong:

student(name="Ali", 20)

✅ Correct:

student("Ali", age=20)

👉 Keyword must come after positional

### Real Life Use

In Data Science & AI:

Functions have many inputs\
Keyword arguments make code clear + readable
