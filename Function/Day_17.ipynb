# **Recursion (Function Calling Itself)**

### Concept Explanation

👉 Recursion = function calling itself

💡 Instead of loop (for, while)
We use function again and again

### Story Time (Mirror Reflection)

Imagine you stand between mirrors 🪞🪞

You see:

- image inside image
- again inside
- again inside

👉 It repeats again and again

💡 That is recursion

### Very Important Rule

👉 Every recursion MUST have a STOP condition

👉 This is called:

👉 Base Case

Without base case ❌ → infinite loop → crash 💥

### Python Simple Example

```python
def countdown(n):
    if n == 0:          # base case
        print("Stop")
    else:
        print(n)
        countdown(n-1)  # function calling itself
```

```python
countdown(5)
```

### Factorial Example

5! = 5 × 4 × 3 × 2 × 1

```python
def factorial(n):
    if n == 1:              # base case
        return 1
    else:
        return n * factorial(n-1)
```

```python
print(factorial(5))
```

### Step Thinking

factorial(5)\
= 5 × factorial(4)\
= 5 × 4 × factorial(3)\
= ...\
= 120

### Sum of Numbers

```python
def total(n):
    if n == 0:
        return 0
    else:
        return n + total(n-1)
```

```python
print(total(5))
```
