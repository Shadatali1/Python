### Python Example

```python
def make_tea():
    print("Tea is ready ☕")
```

👉 This only creates function

**Now we CALL it:**

```python
make_tea()
```

### Call Multiple Times

```python
make_tea()
make_tea()
make_tea()
```

**One function → many uses**
