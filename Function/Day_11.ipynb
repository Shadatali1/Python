# Scope (Local vs Global)

Simple Meaning

👉 Scope = where a variable can be used

💡 Think:
“Where can I access this variable?”

### Story Time

Imagine a house:

🛏️ Your room → local\
🏠 Whole house → global

👉 Things in your room → only you use\
👉 Things in house → everyone can use

### 1. Local Variable

👉 Created inside function

```python
def my_func():
    x = 10
    print(x)
```

✅Work

```python
my_func()
```


👉 ❌ Error:

```python
print(x)   # outside function
```

💡 **Because x is local**

### 2. Global Variable

👉 Created outside function

```python
x = 20

def my_func():
    print(x)
```

👉 Works:

```python

my_func()
print(x)
```


💡 **Global = usable everywhere**

### Important Rule

👉 Inside function → Python first checks local\
👉 If not found → checks global

### Python Example

```python
x = 50

def test():
    x = 10
    print("Inside:", x)

test()
print("Outside:", x)
```

### Using global Keyword

👉 If you want to change global variable inside function

```python
x = 5

def change():
    global x
    x = 10

change()
print(x)
```

### Real Life Example
| Type   | Example              |
| ------ | -------------------- |
| Local  | Your pocket money 💰 |
| Global | Family money 🏠      |
